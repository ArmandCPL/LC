```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,14,100)
plt.close("all")

## Import all the necessary libraries from Python
```

```python
##plot the figure
plt.figure(2,figsize=(12,8))
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 
plt.plot([0,2.77],[1.5,1.5],'b')
plt.plot([2.77,2.77],[1.5,2],'b')
plt.plot([2.77,8.65],[1.5,0.44],'b')
plt.plot([0,8.65],[-1.17,-1.17],'b')
plt.plot([8.65,14],[0.44,0.119],'b')
plt.plot([8.65,14],[-1.17,-1.49],'b')
plt.plot([8.65,8.65],[0.44,-1.17],'b')
plt.text(1,1.6,r'Mn$^{3+}$',weight = 'bold', fontsize=19,color='b')
plt.text(6,1.1,r'Mn(OH)$_{3(s)}$',fontsize=19,color='b')
plt.text(1.5,0,r'Mn$^{2+}$',fontsize=19,color='b')
plt.text(10,-0.4,r'Mn(OH)$_{2(s)}$',fontsize=19,color='b')
plt.text(6,-1.5,r'Mn$_{(s)}$',fontsize=19,color='b')
##Lignes du diiode
#-----------------
plt.plot([0,6.9],[0.68,0.68],'r')
plt.plot([0,6.9],[1.17,0.68],'r')
plt.plot([6.9,14],[0.68,0.254],'r')
plt.text(2,0.8,r'I$_2$',fontsize=19,color='r')
plt.text(4,0,r'I$^{-}$',weight = 'bold', fontsize=19,color='r')
plt.text(8,1.5,r'IO$^{−}_3$',fontsize=19,color='r')
#-----------------
##lignes de l'eau
#-----------------
plt.plot(x,1.23-0.06*x,'k--') ## Oxygen frontier with water
plt.plot(x,-0.06*x,'k--')     ## Hydrogen frontier with water
plt.text(10.4,0.85,r'O$_{2(g)}$', fontsize=19) ## Oxygen label
plt.text(7,-0.8,r'H$_{2(g)}$',fontsize=19) ## Hydrogen label
plt.text(6.3,0,r'H$_{2}$O$_{(g)}$',fontsize=22) ## water label
#----------------------
plt.xlim(0,14)
#plt.title(r'MnC$_0$=10$^{−2}$')  #if needed add concentration of manganese
plt.xlabel('pH', fontsize=22)
plt.ylabel('E(V)', fontsize=22)
plt.savefig('diagramme_mn_eau.png',dpi=300)
```
